# 1. Setups 

```python
import os
import requests
import zipfile
from tqdm.notebook import tqdm
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed

# --- Configuration ---
save_dir = Path("../dataset/cfd_bench/cylinder")
save_dir.mkdir(parents=True, exist_ok=True)

# The specific URLs you provided
urls = {
    "bc.zip": "https://huggingface.co/datasets/chen-yingfa/CFDBench/resolve/main/cylinder/bc.zip",
    "geo.zip": "https://huggingface.co/datasets/chen-yingfa/CFDBench/resolve/main/cylinder/geo.zip",
    "prop.zip": "https://huggingface.co/datasets/chen-yingfa/CFDBench/resolve/main/cylinder/prop.zip"
}

# --- Worker Function ---
def get_chunk(args):
    url, start, end, part_name = args
    headers = {'Range': f'bytes={start}-{end}'}
    try:
        # allow_redirects=True is required for Hugging Face
        r = requests.get(url, headers=headers, stream=True, timeout=60, allow_redirects=True)
        with open(part_name, "wb") as f:
            f.write(r.content)
        return part_name
    except Exception as e:
        return None

# --- Main Execution ---
num_workers = 16

print(f"ðŸš€ Starting Parallel Download to {save_dir}...\n")

for filename, url in urls.items():
    final_path = save_dir / filename
    
    # 1. Check if zip already exists
    if final_path.exists():
        print(f"âœ… {filename} already exists. Skipping download.")
    else:
        print(f"ðŸ“¥ Processing {filename}...")
        try:
            # Get Size
            resp = requests.head(url, allow_redirects=True)
            file_size = int(resp.headers.get('content-length', 0))
            chunk_size = file_size // num_workers
            
            # Create Tasks
            tasks = []
            for i in range(num_workers):
                start = i * chunk_size
                end = file_size - 1 if i == num_workers - 1 else (i + 1) * chunk_size - 1
                part_name = final_path.with_suffix(f".part{i}")
                tasks.append((url, start, end, part_name))
                
            # Run Download
            with ThreadPoolExecutor(max_workers=num_workers) as pool:
                list(tqdm(pool.map(get_chunk, tasks), total=num_workers, desc=f"Downloading {filename}"))

            # Stitch
            print(f"   ðŸ§© Stitching parts...")
            with open(final_path, 'wb') as outfile:
                for i in range(num_workers):
                    part_name = final_path.with_suffix(f".part{i}")
                    if part_name.exists():
                        with open(part_name, 'rb') as infile:
                            outfile.write(infile.read())
                        part_name.unlink() # Cleanup part file
            print(f"   âœ… Downloaded {filename}")
            
        except Exception as e:
            print(f"   âŒ Failed to download {filename}: {e}")
            continue

    # 2. Extraction (Auto-Unzip)
    # FIX: Use final_path.stem (which is a Path object) instead of filename.stem
    extract_dir = save_dir / final_path.stem  
    
    if extract_dir.exists() and any(extract_dir.iterdir()):
        print(f"   â­ï¸  {extract_dir.name} already extracted. Skipping.")
    else:
        print(f"   ðŸ“¦ Extracting to {extract_dir}...")
        try:
            os.makedirs(extract_dir, exist_ok=True)
            with zipfile.ZipFile(final_path, 'r') as zip_ref:
                zip_ref.extractall(extract_dir)
            print(f"   âœ¨ Extracted!")
        except Exception as e:
            print(f"   âŒ Extraction failed: {e}")

print("\nðŸŽ‰ All tasks complete. Data is ready for training.")
```
